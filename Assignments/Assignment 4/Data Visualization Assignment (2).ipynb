## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,10000)
plt.plot(x,np.sin(x),linestyle='-.',color='black')
plt.grid(True)
plt.ylabel('y',size=25)
plt.xlabel('x',size=25)
plt.title("The Sin Curve")
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accesories=np.array([15,20,14,13,9])
stores=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
ind = [x for x, _ in enumerate(stores)]
plt.bar(ind, clothing, width=0.8, label='clothing', color='red', bottom=Equipment+Accesories)
plt.bar(ind, Equipment, width=0.8, label='Equipment', color='blue', bottom=Accesories)
plt.bar(ind, Accesories, width=0.8, label='Accesories', color='green')
plt.xticks(ind, stores)
plt.ylabel("quanities")
plt.xlabel("Cities")
plt.legend(loc="upper right")
plt.grid(False)
plt.show()
```
